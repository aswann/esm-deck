```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr

```

```python
t = np.arange(350 * 12) / 12.
y = np.zeros(350 * 12) * 1.
y[0:1200] = 10.
y[1200:2400] = 10. - (0.2 *  np.arange(100*12)/12.)
y[2400:3601] = -10.


yy = y.cumsum()/12.

#plt.plot(t,y)

t2 = np.arange(170 * 12) / 12.
y2 = np.ones(170 * 12) * 10.
y2y2 = y2.cumsum()/12.

#plt.plot(t2,y2)
```

```python


width=2.5
symcolor='green'
symcolor2 = 'blue'
linecolor2 = 'green'

fontsize=15

plt.rc('xtick', labelsize=fontsize) 

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(t2,y2y2, color=linecolor2, linestyle='--')

ax.plot(t,yy, color='navy', linestyle='--')

ax.plot(0,0, '_', color=symcolor, markersize=20, markeredgewidth=width)
ax.annotate('T$_{0}$', (0, 0), xytext=(10,0), textcoords='offset points', color=symcolor, fontsize=fontsize)

ax.plot(300,0, '_', color=symcolor, markersize=20, markeredgewidth=width)
ax.annotate('T$_{300}$', (300, 0), xytext=(-50,0), textcoords='offset points', color=symcolor, fontsize=fontsize)

ax.plot(150,1250, '|', color=symcolor2, markersize=50, markeredgewidth=width)
ax.annotate('t$_{peak}$', (150,1250), xytext=(10,30), textcoords='offset points', color=symcolor2, fontsize=fontsize)

ax.plot(100,1000, '_', color=symcolor, markersize=20, markeredgewidth=width)
ax.annotate('T$_{100}$', (100, 1000), xytext=(10,0), textcoords='offset points', color=symcolor, fontsize=fontsize)

ax.plot(200,1000, '_', color=symcolor, markersize=20, markeredgewidth=width)
ax.annotate('T$_{200}$', (200, 1000), xytext=(10,0), textcoords='offset points', color=symcolor, fontsize=fontsize)

ax.plot(125,1250, '_', color=symcolor, markersize=20, markeredgewidth=width)
ax.annotate('T$_{ref}$', (125, 1250), xytext=(-40,0), textcoords='offset points', color=symcolor, fontsize=fontsize)

ax.plot(150,1250, '_', color=symcolor, markersize=20, markeredgewidth=width)
ax.annotate('T$_{150}$', (150, 1250), xytext=(10,0), textcoords='offset points', color=symcolor, fontsize=fontsize)

plt.ylabel('Hypothetical GMST [K]', fontsize=fontsize)
plt.xlabel('Simulation years', fontsize=fontsize)
plt.yticks([])

ax.text(230, 1650, r'TNZ = T$_{150}$ - T$_{ref}$', fontsize=fontsize)
ax.text(230, 1550, r'TR1000 = T$_{200}$ - T$_{100}$', fontsize=fontsize)
ax.text(230, 1450, r'TR0 = T$_{300}$ - T$_{0}$', fontsize=fontsize)
ax.text(230, 1350, r'tPW = t$_{peak}$ - 150', fontsize=fontsize)

plt.savefig('images/metrics_schematic.png', dpi=300)
```
