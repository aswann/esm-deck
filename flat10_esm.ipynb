```python
mdls=['ACCESS-ESM1-5','CESM2','GFDL-ESM4','NASA-GISS','NorESM2-LM']
```

```python
expt=['esm-pictrl','1pctco2','flat10','flat10-zec','flat10-cdr']
```

```python
flds=['tas','co2','nbp','fgco2']
```

```python
ed={}
for m in mdls:
    ed[m]={}
    for e in expt:
        tmpa=[]
        for f in flds:
            try:
                fl='ESM_data/'+m+'/'+e+'_'+f+'_'+m+'.nc'
                tmp=xr.open_dataset(fl)
                tmpa.append(tmp)
            except:
                0
        ed[m][e]=xr.merge(tmpa)
                
                
```
